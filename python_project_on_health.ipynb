```python
# function for treating the outliers
def remove_outlair(df1,boxplot):
  q1=df1['Adult_Mortality'].quantile(0.25)
  q3=df1['Adult_Mortality'].quantile(0.75)

  iqr=q3-q1

  lower_bond=q1-1.5*iqr
  upper_bond=q3+1.5*iqr

  outlairs=df1.index[(df1['Adult_Mortality']<lower_bond)| (df1['Adult_Mortality']>upper_bond)]
  final_df=df1.drop(outlairs)

  return final_df

df1=remove_outlair(df1,'Adult_Mortality')
```
